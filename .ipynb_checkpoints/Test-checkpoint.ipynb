```python

from graph_tool.all import *
from makegraph import make_graph
from numpy.random import randint
from numpy.random import random
from numpy.random import *
import numpy as np
```

```python
g=make_graph()
#zero matrix 100x5
matrix = np.zeros([100,5])
```

```python
#states
S = [1, 1, 1, 1]           

I = [0, 0, 0, 1]           

R = [0.5, 0.5, 0.5, 1.]    

V = [0, 0, 1, 1]           

D = [0.8, 0, 0, 0.6] 
```

```python
#initialization of matrix, every row encodes state in one hot manner, e.g. [1,0,0,0,0] is S, [0,1,0,0,0] is I
counter = 0

for v in g.vertices():
    if g.vp.state[v] == S: 
        matrix[counter,0] = 1
    if g.vp.state[v] == I: 
        matrix[counter,1] = 1
    if g.vp.state[v] == R: 
        matrix[counter,2] = 1
    if g.vp.state[v] == V: 
        matrix[counter,3] = 1
    if g.vp.state[v] == D: 
        matrix[counter,4] = 1
    counter += 1
print(matrix)
```
